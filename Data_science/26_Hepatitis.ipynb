```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
df = pd.read_csv("../DSBDAL_Exam_DataSets/Hepatitis/hepatitis.csv", header=None)
df
```

```python
df.columns = ["Class","Age","Sex","Steroid","Antivirals","Fatigue","Malaise","Anorexia","Liver Big","Liver Firm","Spleen","Spiders","Ascites","Varices","Bilirubin","Alk Phosphate","Sgot","Albumin","Protime","Histology"]
df
```

```python
df.info()
```

```python
df.shape
```

# Data Cleaning

```python
df.isna().sum()
```

```python
df.replace('?',np.nan,inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.apply(pd.to_numeric,errors='ignore')
df
```

```python
df.dtypes
```

```python
df=df.astype(float)
```

# Outliers

```python
df.shape
```

```python
out = abs((df - df.mean()) / df.std())
lier = out > 3
lier.sum()
```

```python
df = df[~lier.any(axis=1)]
df
```

# Data Transformation

```python
num_cols = df.select_dtypes(include=np.number).columns
```

```python
scaler=StandardScaler()
df[num_cols]=scaler.fit_transform(df[num_cols])
df
```

```python
df['Class']=np.where(df['Class']>=df['Class'].median(),1,0)
```

# Model

```python
X = df.drop('Class',axis=1)
Y = df['Class']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
pred_lr = lr.predict(X_test)
pred_lr
```

```python
pred_lr_acc = accuracy_score(Y_test, pred_lr)
pred_lr_acc
```

# Naive Bayes

```python
g_NB = GaussianNB()
g_NB.fit(X_train, Y_train)
```

```python
pred_gnb = g_NB.predict(X_test)
pred_gnb
```

```python
pred_gnb_acc = accuracy_score(Y_test, pred_gnb)
pred_gnb_acc
```
