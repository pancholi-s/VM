# Create data subsets for type of post

```python
df_photo = df[df["Type"]=="Photo"]
df_photo.head()
```

```python
df_status = df[df["Type"]=="Status"]
df_status.head()
```

```python
df_link = df[df["Type"]=="Link"]
df_link.head()
```

```python
df_video = df[df["Type"] == "Video"]
df_video.head()
```

# Sort Data on Page Total Likes

```python
df_sort_likes = df.sort_values(by="like")
df_sort_likes
```

```python
df_sort_comments = df.sort_values(by="comment", ascending=False)
df_sort_comments
```

# Transposing

```python
df_transpose = df.transpose()
df_transpose
```

# Shape and Reshape

```python
# adding an index column to provide a unique value to each row entry to further melt and unmelt the data

df["index"] = range(1, len(df)+1)
df.head()
```

# Melting data to long format

```python
df_melt = pd.melt(df, id_vars=['Type', 'Category'], var_name='Metrics', value_name='Value')
df_melt
```

```python
df_melt = pd.melt(df, id_vars='index', value_vars='Total Interactions')
df_melt
```

```python
df_melt = pd.melt(df, id_vars='index', value_vars='Type')
df_melt
```

# Casting to wide format

```python
df_cast = pd.pivot_table(df, index=['Type', 'Category'], values='like', aggfunc=list)
df_cast
```
