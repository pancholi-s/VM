```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DSBDAL_Exam_DataSets/heart.csv')
df
```

```python
df.isna().sum()
```

```python
(df=='?').sum().sum()
```

```python
df = df.replace('?', np.nan)
df = df.dropna()
```

```python
df.shape
```

# Histogram

## Objective: Visualize the distribution of a continuous variable, such as age or cholesterol levels, in the dataset.

```python
cols = ['age', 'chol', 'thalach']
df[cols].hist(bins=10, figsize=(15,10))
plt.tight_layout()
```

# Dot

## Objective: Compare the distribution of a categorical variable, such as chest pain type or presence of fasting blood sugar, with respect to another variable.

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='cp', y='trestbps', data=df, palette='Set2', hue='target')
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting BP')
plt.title('Chest Pain vs. Resting Blood Pressure')
plt.show()
```

# BAR
## Objective: Compare the values of a categorical variable, such as the frequency of different types of heart disease, or the distribution of sexes.

```python
plt.figure(figsize=(8,6))
df['target'].value_counts().plot(kind='bar', color='green')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.title('Heart Disease')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['sex'].value_counts().plot(kind='bar', color='red')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Sex Distribution')
plt.show()
```

# Line

## Objective: Visualize trends or patterns in a continuous variable, such as heart rate (thalach) over time.

```python
plt.figure(figsize=(10,6))
plt.plot(df['thalach'])
plt.xlabel('Index')
plt.ylabel('Heart-Rate')
plt.title('Heart-rate vs Index')
plt.show()
```

# Scatter Plot

## Objective: Identify relationships or correlations between two continuous variables, such as age and cholesterol levels, and explore any potential patterns

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='age', y='chol', hue='thal', palette='Set2')
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.show()
```

```python
#subplots
fig, axes = plt.subplots(1, 3, figsize=(18,6))

#boxplot
sns.boxplot(data=df, x='ca', y='age', ax=axes[0])
axes[0].set_title("Box plot of ca vs age")

#histogram
sns.histplot(data=df, x='chol', kde=True, ax=axes[1])
axes[1].set_title("Histogram of chol")

#scatterplot
sns.scatterplot(data=df, x='age', y='trestbps', hue='target', ax=axes[2])
axes[2].set_title("Scatter - Age vs Resting BP")

plt.tight_layout()
plt.show()
```

# Pie Charts

## Objective: Compare the distribution of sexes in the dataset

```python
plt.figure(figsize=(8,6))
df['sex'].value_counts().plot(kind='pie', autopct='%1.2f%%', colors=['blue', 'red'], labels=['male', 'female'])
plt.title('Sex')
plt.show()
```

# Box Plots

## Objective: Compare cholesterol levels for different chest pain types

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='cp', y='chol', data=df)
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholesterol')
plt.title('Chest Pain vs. Cholesterol')  
plt.show()
```

```python
#subplots
fig, axes = plt.subplots(1, 2, figsize=(18,6))

#scatterplot
sns.scatterplot(data=df, x='age', y='trestbps', hue='target', ax=axes[0])
axes[0].set_title("Scatter - Age vs Resting BP")

#boxplot
sns.boxplot(data=df, x='ca', y='age', ax=axes[1])
axes[1].set_title("Box plot of ca vs age")

plt.tight_layout()
plt.show()
```
