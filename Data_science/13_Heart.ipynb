```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("../DSBDAL_Exam_DataSets/heart.csv")
df
```

# Data Cleaning

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df1 = df.replace('?', pd.NA)
df1
```

```python
df1.isna().sum()
```

```python
df = df1[df1 >= 0]
df
```

# Error Correcting (outliers)

```python
#standard deviation

df.std()
```

```python
df.mean()
```

```python
out = np.abs((df - df.mean()) / df.std())
lier = out>3
lier.sum()
```

```python
df = df[~lier.any(axis=1)]
df
```

# Data Transformation

```python
df.loc[df['sex'] == 1, 'sex'] = 'M'
df.loc[df['sex'] == 0, 'sex'] = 'F'
df
```

```python
lb = LabelEncoder()
df['sex'] = lb.fit_transform(df['sex'])
df
```

# Model - Regression, KNN

```python
X = df.drop('target', axis=1)
X
```

```python
Y = df['target']
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=46)
```

# Logistic Regression

```python
lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, Y_train)
```

```python
pred_lr = lr.predict(X_test)
pred_lr
```

```python
pred_lr_acc = accuracy_score(Y_test, pred_lr)
pred_lr_acc
```

# KNN

```python
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
```

```python
pred_knn = knn.predict(X_test)
pred_knn
```

```python
pred_knn_acc = accuracy_score(Y_test, pred_knn)
pred_knn_acc
```
