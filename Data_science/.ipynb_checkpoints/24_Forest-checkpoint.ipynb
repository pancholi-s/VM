# Create Subsets

```python
df['area'].mean()
```

```python
notAffected = df[df['area']==0]
notAffected
```

```python
mostlyAffected = df[df['area'] > 13 ]
mostlyAffected
```

```python
partiallyAffected = df[(df['area']  <= 13 ) & (df['area']  > 0 )]
partiallyAffected
```

# Merge

```python
df_merge1 = pd.concat([notAffected, partiallyAffected])
df_merge1
```

```python
df_merge2 = pd.concat([mostlyAffected, partiallyAffected])
df_merge2
```

# Melting

```python
df_melt = pd.melt(df, id_vars="area", value_vars="temp")
df_melt
```

# Casting

```python
df_cast = pd.pivot_table(df, index=['temp', 'rain'], values='area', aggfunc=list)
df_cast
```
