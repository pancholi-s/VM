```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../DSBDAL_Exam_DataSets/Iris/Iris.csv", header=None)
df
```

```python
df.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
df
```

```python
df.shape
```

```python
df.info()
```

# Create subsets

```python
df['Species'].unique()
```

```python
df_setosa = df[df['Species'] == 'Iris-setosa']
df_setosa.head()
```

```python
df_versicolor = df[df['Species'] == 'Iris-versicolor']
df_versicolor.head()
```

```python
df_virginica = df[df['Species'] == 'Iris-virginica']
df_virginica.head()
```

# Merge 2 subsets

```python
df_merge = pd.concat([df_setosa, df_versicolor])
df_merge
```

# Sort on petal length

```python
df_sort = df.sort_values('PetalLength')
df_sort
```

```python
df_sort = df.sort_values('PetalLength', ascending=False)
df_sort
```

# Transpose

```python
df_transpose = df.transpose()
df_transpose
```

# Melting to long form

```python
df_melt = pd.melt(df, id_vars=['SepalLength', 'PetalLength', 'Species'])
df_melt
```

```python
df_melt = pd.melt(df, id_vars='Species', var_name='Measures')
df_melt
```

# Casting to wide form

```python
df_cast = pd.pivot_table(df, index='Species', aggfunc=list)
df_cast.head()
```
