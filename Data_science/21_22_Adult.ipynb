```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DSBDAL_Exam_DataSets/Adult/adult.csv')
df
```

```python
df.isna().sum()
```

```python
(df=='?').sum().sum()
```

```python
df = df.replace('?', np.nan)
df = df.dropna()
```

```python
df.shape
```

# Histogram

## Objective: Visualize the distribution of a continuous variable, such as age or cholesterol levels, in the dataset.

```python
col=['Age','Education Num','Hours Per Week']
df[col].hist(bins=10,figsize=(10,6))
plt.tight_layout()
```

# Dot

## Objective: Compare the distribution of a categorical variable, such as chest pain type or presence of fasting blood sugar, with respect to another variable.

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='Workclass', y='Education', data=df, palette='Set2', hue='Income')
plt.xlabel('Work Class')
plt.ylabel('Education')
plt.title('Work Class vs. Education')
plt.show()
```

# BAR

## Objective: Compare the values of a categorical variable, such as the frequency of different types of heart disease, or the distribution of sexes

```python
plt.figure(figsize=(15,6))
plt.bar(df['Workclass'].value_counts().index, df['Workclass'].value_counts().values)
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Distribution of Workclass')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.bar(df['Education'].value_counts().index, df['Education'].value_counts().values)
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Distribution of Workclass')
plt.show()
```

# Line

## Objective: Visualize trends or patterns in a continuous variable, such as heart rate (thalach) over time.

```python
plt.figure(figsize=(18,6))
plt.plot(df['Age'])
plt.xlabel('Age')
plt.ylabel('Index')
plt.title('Age vs Index')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.plot(df['Education Num'])
plt.xlabel('Education Num')
plt.ylabel('Index')
plt.title('Education Num vs Index')
plt.show()
```

# Scatter Plot

## Objective: Identify relationships or correlations between two continuous variables, such as age and cholesterol levels, and explore any potential patterns

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Workclass', y='FnlWgt', hue='Income', palette='Set2')
plt.xlabel('Work Class')
plt.ylabel('Number')
plt.show()
```

```python
#subplots
fig, axes = plt.subplots(1, 3, figsize=(18,6))

#boxplot
sns.boxplot(data=df, x='Hours Per Week', y='Education Num', ax=axes[0])
axes[0].set_title("Box plot of ca vs age")

#histogram
sns.histplot(data=df, x='Age', kde=True, ax=axes[1])
axes[1].set_title("Histogram of Age")

#scatterplot
sns.scatterplot(data=df, x='Workclass', y='FnlWgt', hue='Income', palette='Set2', ax=axes[2])
axes[2].set_title("Scatter - Work Class vs FnlWgt")

plt.tight_layout()
plt.show()
```

# Pie Charts

## Objective: Compare the distribution of sexes in the dataset

```python
# Objective: Illustrate the proportion of different races using a pie chart
plt.pie(df['Race'].value_counts().values, labels=df['Race'].value_counts().index, autopct='%1.1f%%')
plt.title('Race Composition')
plt.show()
```

# Box Plots

## Objective: Compare cholesterol levels for different chest pain types

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Hours Per Week', y='Education Num', data=df)
plt.xlabel('Hours Per Week')
plt.ylabel('Education Num')
plt.title('Hours per Week vs. Education Num')  
plt.show()
```

```python
#subplots
fig, axes = plt.subplots(1, 2, figsize=(18,6))

#scatterplot
sns.scatterplot(data=df, x='Workclass', y='FnlWgt', hue='Income', palette='Set2', ax=axes[0])
axes[0].set_title("Scatter - Work Class vs FnlWgt")

#boxplot
sns.boxplot(x='Hours Per Week', y='Education Num', data=df, ax=axes[1])
axes[1].set_title("Box plot of hrs/week vs Education Num")

plt.tight_layout()
plt.show()
```
